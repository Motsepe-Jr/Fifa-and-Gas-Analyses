```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

plt.figure(figsize =(8, 5))

plt.style.use('ggplot')
gas = pd.read_csv('gas_prices.csv')
# gas.head()

plt.plot(gas.Year, gas.USA, 'g.--', label = 'USA')
plt.plot(gas.Year, gas.Canada, 'ro-',label = 'Canada') # You can also use gas[""]

# # use a list to grap specifi things
# countries_to_look_at = ['USA', 'Canada', 'South Korea', 'Australia']
# for country in gas:
#     if country in countries_to_look_at:
#         plt.plot(gas.Year, gas[country], marker='o', label ='%s data' %country)

plt.title('Gas Price Over Time (USD)', fontdict = {'fontweight': 'bold', 'fontsize':22})

plt.xticks(gas.Year[::3].tolist()+[2011]) # or use gas.year[::3]
plt.legend()
plt.xlabel('Years')
plt.ylabel('US Dollars $')
plt.show()
```

### Load Fifa data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head(10)
```

### Histogram

```python
bins = [ 40, 50, 60,70, 80,90, 100]

plt.figure(figsize=(9,7))
plt.hist(fifa.Overall, bins =bins)

plt.xticks(bins)
plt.xlabel('Skill level')
plt.ylabel('Number of Players')
plt.title('Distribution of Players Sklls In Fifa (2018)')

plt.show()
```

```python
# Pie Charts
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
labes = ['left', 'Right']
# color = ['red', 'g']
plt.pie([left, right], labels = labes, autopct ='%.2f%%', shadow = True)
plt.title('Foot Preference of Fifa Players')

plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight <125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight <150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight <175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight <200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']

plt.pie(weights, labels = labels, autopct = '%.2f %%', pctdistance =0.8, explode = [0.4, 0.3, 0.0, 0.0, 0.4])
plt.title('Weight Distribution of Fifa Players (Ibs)')
plt.show()
```

```python
plt.style.use('default')

plt.figure(figsize=(8, 6))

labels = ['FC Barcelona', 'Real Madrid', 'New England Revolution']

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
real_madrid = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
Revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall'] 


boxes = plt.boxplot([barcelona, real_madrid, Revs], labels = labels, patch_artist = True, medianprops={'linewidth':2})

for box in boxes['boxes']:
    # set edge color
    box.set(color='red', linewidth=2)
    #change fill color
    box.set(facecolor = '#e0e0e0')

plt.title('Professional Football Team Comparison (Fifa)')
plt.ylabel('Fifa Overall Ratings')

plt.show()
```
